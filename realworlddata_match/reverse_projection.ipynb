```python
import numpy as np
```

```python
H = np.array([[ 1.01117006e+00, -9.98803211e-02, -5.84929931e-03,
         1.00240922e+02],
       [-1.05018563e-02,  6.83070936e-01, -5.92992245e-01,
         2.49459026e+01],
       [ 2.18134506e-05, -1.64264098e-04, -6.50189847e-05,
         1.08401770e+00]])
```

```python
def antisa(x,y):
    U = np.dot(H,np.array([x,y,1.7,1]))
    return (3*int(U[0]/U[2]),3*int(U[1]/U[2]))
```

```python
import cv2
```

```python
cap = cv2.VideoCapture('./realworld.mp4')
```

```python
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))
```

```python
output_path = "output.mp4"
out = cv2.VideoWriter(output_path, cv2.VideoWriter_fourcc(*'mp4v'), fps, (frame_width, frame_height))
```

```python
data = []
with open('./res/run_standard.txt','r') as file:
    data = [_.split(' ') for _ in file.read().split('\n')]
data
```

```python
t = 0

pointer = 0

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    t+=1

    if t%100 == 0:
        print(f'frame {t*10}')

    while int(data[pointer][1]) == t:
        radius = 30
        thickness = -1
        c = (int(data[pointer][2].split('.')[0]),int(data[pointer][3].split('.')[0])+100)
        coord = antisa(c[0],c[1])
        cv2.circle(frame, (coord[0],coord[1]-300), radius, (0,255,0), thickness)
        pointer += 1
        if pointer==len(data):
            pointer -= 1

    cv2.line(frame, antisa(0,100), antisa(1200,100), (0,0,255), 5)
    cv2.line(frame, antisa(1200,100), antisa(1200,838), (0,0,255), 5)
    cv2.line(frame, antisa(1200,838), antisa(0,838), (0,0,255), 5)
    cv2.line(frame, antisa(0,838), antisa(0,100), (0,0,255), 5)

    out.write(frame)
```

```python
cap.release()
out.release()  
cv2.destroyAllWindows()
```
